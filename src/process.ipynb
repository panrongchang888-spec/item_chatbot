```python
token = 100
intro = data['intro'].to_list()[0]
#print(intro)
print(len(intro))
result_list = []
if len(intro) % 100 == 0:
    block = len(intro) // 100
else:
    block = len(intro) // 100 + 1
with open('./data/test.txt', 'w') as f:
   
    for i in range(block):
        start = i * token
        end = (i + 1) * token
        if end > len(intro):
            end = len(intro)
        print(f'block {i+1}: from {start} to {end}')
        print(intro[start:end])
        result_list.append(intro[start:end])
        f.write(intro[start:end] + '\n')
f.close()
```

```python
from datasets import Dataset
texts = [
    "この商品は返品できますか？",   # 退货
    "この商品の値段はいくらですか？", # 商品咨询
    "こんにちは！",                 # 闲聊
    "明日の天気は？"                # 非相关
]
labels = [0, 1, 2, 3]
dataset = Dataset.from_dict({"text": texts, "label": labels})
dataset = dataset.train_test_split(test_size=0.2)
```

```python
dataset['train'][0]
```

```python
import torch
train_loader = torch.utils.data.DataLoader(dataset["train"], batch_size=2, shuffle=True)
```
